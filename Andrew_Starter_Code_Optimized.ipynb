## Preprocessing

# import keras added to help with saving weights

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

import keras

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_cleaned = application_df.drop('EIN', axis = 1)
application_df_cleaned

```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
application_df_cleaned.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
application_df_cleaned['NAME'].value_counts()
```

```python
name_types_to_replace=application_df_cleaned['NAME'].value_counts()
name_types_to_replace=name_types_to_replace[name_types_to_replace<10].index
name_types_to_replace
```

```python
for app in name_types_to_replace:
    application_df_cleaned['NAME'] = application_df_cleaned['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_df_cleaned['NAME'].value_counts()
```

#APPLICATION TYPE STARTS HERE


```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
application_df_cleaned['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace=application_df_cleaned['APPLICATION_TYPE'].value_counts()
application_types_to_replace=application_types_to_replace[application_types_to_replace<500].index
application_types_to_replace
```

```python

# Replace in dataframe
for app in application_types_to_replace:
    application_df_cleaned['APPLICATION_TYPE'] = application_df_cleaned['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_cleaned['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
application_df_cleaned['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classification_above_1=application_df_cleaned['CLASSIFICATION'].value_counts()
classification_above_1=classification_above_1[classification_above_1>1]
classification_above_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace=application_df_cleaned['CLASSIFICATION'].value_counts()
classifications_to_replace=classifications_to_replace[classifications_to_replace<1800].index
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_cleaned['CLASSIFICATION'] = application_df_cleaned['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df_cleaned['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df_cleaned)
print(application_df_dummies.columns)
application_df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
Features = application_df_dummies.drop('IS_SUCCESSFUL',axis = 1).values
Target = application_df_dummies['IS_SUCCESSFUL'].values
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Features, Target, random_state=46)
```

```python

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# nn_model = tf.keras.models.Sequential()

# # Add our first Dense layer, including the input layer
# nn_model.add(tf.keras.layers.Dense(units=1, activation="relu", input_dim=2))

# # Add the output layer that uses a probability activation function
# nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# # Check the structure of the Sequential model
# nn_model.summary()
# features count for Input Nodes
input_nodes = len(X_train[0])
input_nodes
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
input_nodes = len(X_train[0])
first_nodes = 20
sec_nodes = 10


nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=first_nodes, activation="relu", input_dim=input_nodes))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=sec_nodes, activation="sigmoid"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn_model.fit(X_train_scaled,y_train,epochs=20, callbacks=[mc])
# fit_model = nn_model.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
